# Machine Learning

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
con.Gender = le.fit_transform(con.Gender)
con.Item_purchased = le.fit_transform(con.Item_purchased)
con.Location=le.fit_transform(con.Location)
con.Size=le.fit_transform(con.Size)
con.Color=le.fit_transform(con.Color)
con.Season=le.fit_transform(con.Season)
con.Subscription_Status=le.fit_transform(con.Subscription_Status)
con.Shipping_Type=le.fit_transform(con.Shipping_Type)
con.Discount_Applied=le.fit_transform(con.Discount_Applied)
con.Promo_Code_Used=le.fit_transform(con.Promo_Code_Used)
con.Payment_Method=le.fit_transform(con.Payment_Method)
con.Frequency_of_Purchases=le.fit_transform(con.Frequency_of_Purchases)
```

```python
ip=con.drop(['Subscription_Status','Category'],axis=1)
op=con.Subscription_Status
```

```python
ip,op
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xts,ytr,yts = train_test_split(ip,op,test_size=0.2,random_state=7)
```

```python
xtr
```

```python
print(xtr.size,xts.size)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtr1 = sc.fit_transform(xtr)
xts1= sc.fit_transform(xts)
```

```python
from sklearn.preprocessing import MinMaxScaler
mc = MinMaxScaler()
xtr2 = mc.fit_transform(xtr) 
xts2= mc.fit_transform(xts)
```

```python
print(xtr,xts)
```

```python
ytr.unique()
```

```python
from sklearn.linear_model import LogisticRegression
le=LogisticRegression()
```

```python
le.fit(xtr1,ytr)
```

```python
y_pred = le.predict(xts1)
```

```python
from sklearn.metrics import f1_score,recall_score,precision_score,accuracy_score
acc=accuracy_score(y_pred,yts)
pre=precision_score(y_pred,yts,average='macro')
rec = recall_score(y_pred,yts,average='macro')
f1 = f1_score(y_pred,yts,average='macro')
print('f1_score=',f1,'accuracy_score=',acc,'precision_score=',pre,'recall_score=',rec)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(yts,y_pred)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```
